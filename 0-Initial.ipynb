# Project goal: 
#### The aim of this project is to analyse data of Health Insurance customers and predict whether they would be interested in an additional Vehicle Insurance. 

As a first step, we are interested in doing an initial exploratory analysis of the training data, to see the correlation of the response to the other variables. We start by first reading in all necessary libraries and the training data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Gather

```python
df = pd.read_csv('data/train.csv')
df.head()
```

## Visualize

Let's now look at the correlation of the 'Response' to the other variables. Here we take all variables except for the descriptive ones ('Gender', 'Vehicle_Age', 'Vehicle_Damage'):

```python
df = df.drop(['id'], axis=1) # Drop the 'id' column because it doesn't add to the analysis.

df_corr = df.corr()
sns.heatmap(df_corr, annot = True, fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

The correlation matrix shows us that there is a negative correlation between the 'Response' and the 'Previously_Insured' variables. We will look more into this correlation in the next notebook, where we analyze the age and gender profile of customers.
