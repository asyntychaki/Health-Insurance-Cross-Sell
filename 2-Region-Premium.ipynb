```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
# Total nbr of customers that are interested in a vehicle insurance
df_pos_tot = df[df['Response']==1]
```

```python
# What is the most frequent communication medium of customers interested in an insurance?
df_pos_tot['Policy_Sales_Channel'].value_counts(normalize=True)
```

```python
# What is the most frequent communication medium of customers?
df['Policy_Sales_Channel'].value_counts(normalize=True)
```

```python
(df_pos_tot['Policy_Sales_Channel'].value_counts()/df['Policy_Sales_Channel'].value_counts()).sort_values(ascending=False).plot(kind="bar",figsize=[20,20]);
```

```python
# Do customers who that have a vehicle insurance are older in age?
df_pos_tot.groupby(['Policy_Sales_Channel']).mean()['Annual_Premium'].sort_values(ascending=False)

```
