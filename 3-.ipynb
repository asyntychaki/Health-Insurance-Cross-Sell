```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
plt.figure()
sns.violinplot(data=df, x='Gender', y='Region_Code', hue='Response', palette='Set2', split=True)
plt.show()
```

```python
df_pos = df[df['Response']==1]
df_neg = df[df['Response']==0]

sns.jointplot(x=df_pos['Age'], y=df_pos['Region_Code'], kind="hex", color="#4CB391")
sns.jointplot(x=df_neg['Age'], y=df_neg['Region_Code'], kind="hex", color="#4CB391")
```
